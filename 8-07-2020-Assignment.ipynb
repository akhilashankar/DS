```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=plt.imread("a1.jpg")
print(img.shape)
```

```python
plt.imshow(img)
```

```python
plt.imshow(img[:,:,0])
```

```python
plt.imshow(img[:,:,0],cmap='gray')
```

```python
plt.imshow(img[:,0:600])
```

```python
plt.imshow(img[:,400:800])
```

```python
plt.imshow(img[100:300,200:500])
```
