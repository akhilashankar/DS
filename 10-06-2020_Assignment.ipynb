### Understanding the data
    How many rows are there in the data?
    How many columns are there?
    How many categorical columns are there?
    Print the categorical column names
    Print the numerical column names

```python
print(df.shape[0])
print(df.shape[1])
categorical = df.columns[(df.dtypes.values == np.dtype(object))]
print(categorical)
numerical = df.columns[(df.dtypes.values != np.dtype(object))]
print(numerical)
```

### draw a relplot by palcing views on y and upvotes on x
    draw the plot using data argument
    draw the plot without using data argument

```python
sns.relplot("Upvotes","Views",data=df)
```

```python
sns.relplot(df["Upvotes"],df["Views"])
```

### Add another argument for hue the above plot.
    check it with one numerical column as hue
    check it with one categorical column as hue

```python
sns.relplot("Upvotes","Views",data=df,hue="Answers")
```

```python
sns.relplot("Upvotes","Views",data=df,hue="Tag")
```

### Modify the above plot using any of the parameters size,style,col¶

```python
sns.relplot("Upvotes","Views",data=df,hue="Answers",size=2.5)
```

### Draw the cat plot by taking tag on xaxis and upvotes on yaxis

```python
sns.catplot("Tag","Upvotes",data=df)
```

```python
sns.catplot("Tag","Answers",data=df)
```
